```python
zipcode = pd.DataFrame(Final_foodx['postalCode'].value_counts())
```

```python
zipcode.reset_index(level=0, inplace=True)
```

```python

zipcode.head()
```

```python
zipcodes = zipcode.rename(columns={'index':'ZipCodes', 'postalCode':'Counts'})
```

```python
zipcodes.head()
```

```python
zipcodes.to_csv('newZip.csv')
```

```python
county_geo = r'Capstone Project\nyc_zip_code_tabulation_areas_polygons.geojson'
```

```python
manx= Manhatten
manx.choropleth(
 geo_data=county_geo,
 ##name=food_map,
 data=zipcodes,
 columns=['ZipCodes', 'Counts'],
 key_on='Featuresproperties.postalcode',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Ethnic Restaurants'
)
folium.LayerControl().add_to(manx)
```

```python
food_map = folium.Map(location=[40.74476,-73.965], zoom_start=13) # generate map





for lat, lng, label in zip(Final_foodx.lat, Final_foodx.lng, Final_foodx.categories):
    folium.features.CircleMarker(
        [lat, lng],
        radius=10,
        popup=label,
        fill=True,
        color='blue',
        fill_color='green',
        fill_opacity=0.2
        ).add_to(food_map)

# display map
food_map
```
